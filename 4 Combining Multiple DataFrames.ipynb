## Database-Style DataFrame Merging

### Relationships
| Relationship | Description |
|:-------------|:------------------------------------------------------------------------------------------------|
| One-to-one | Join two columns that contain unique values |
| Many-to-one | Join one column that contains non-unique values with another column that contains unique values |
| Many-to-many | Join two columns that contain non-unique values |

```python
Image("images/merge-one-to-one.png", width=400)
```

```python
# DataFrame merge one-to-one
df_l = pd.DataFrame({
        'A': ['A0', 'A1'],
        'key': ['K0', 'K1']
    })
df_r = pd.DataFrame({
        'B': ['B0', 'B1'],
        'key': ['K0', 'K1']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, on='key')
print(df_merged)
```

```python
Image("images/merge-many-to-one.png", width=400)
```

```python
# DataFrame merge many-to-one
df_l = pd.DataFrame({
        'A': ['A0', 'A1'],
        'key': ['K0', 'K0']
    })
df_r = pd.DataFrame({
        'B': ['B0'],
        'key': ['K0']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, on='key')
print(df_merged)
```

```python
Image("images/merge-many-to-many.png", width=400)
```

```python
# DataFrame merge many-to-many
df_l = pd.DataFrame({
        'A': ['A0', 'A1'],
        'key': ['K0', 'K0']
    })
df_r = pd.DataFrame({
        'B': ['B0', 'B1'],
        'key': ['K0', 'K0']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, on='key')
print(df_merged)
```

### Merge Methods
| Merge Method | Description |
|:-------------|:----------------------------------------------|
| Inner | Use intersection of keys from both DataFrames |
| Outer | Use union of keys from both DataFrames |
| Left | Use keys from left DataFrame only |
| Right | Use keys from right DataFrame only |

```python
Image("images/merge-inner.png", width=400)
```

```python
# DataFrame merge method inner
df_l = pd.DataFrame({
        'A': ['A0', 'A1'],
        'key': ['K0', 'K1']
    })
df_r = pd.DataFrame({
        'B': ['B0', 'B1'],
        'key': ['K0', 'K2']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, on='key') # default how='inner'
print(df_merged)
```

```python
Image("images/merge-outer.png", width=400)
```

```python
# DataFrame merge method outer
df_l = pd.DataFrame({
        'A': ['A0', 'A1'],
        'key': ['K0', 'K1']
    })
df_r = pd.DataFrame({
        'B': ['B0', 'B1'],
        'key': ['K0', 'K2']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, on='key', how='outer')
print(df_merged)
```

```python
Image("images/merge-left.png", width=400)
```

```python
# DataFrame merge method left
df_l = pd.DataFrame({
        'A': ['A0', 'A1'],
        'key': ['K0', 'K1']
    })
df_r = pd.DataFrame({
        'B': ['B0', 'B1'],
        'key': ['K0', 'K2']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, on='key', how='left')
print(df_merged)
```

```python
Image("images/merge-right.png", width=400)
```

```python
# DataFrame merge method right
df_l = pd.DataFrame({
        'A': ['A0', 'A1'],
        'key': ['K0', 'K1']
    })
df_r = pd.DataFrame({
        'B': ['B0', 'B1'],
        'key': ['K0', 'K2']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, on='key', how='right')
print(df_merged)
```

### Merge on Keys (Columns)

```python
# DataFrame merge with different key names (columns)
df_l = pd.DataFrame({
        'A': ['A0', 'A1'],
        'key_l': ['K0', 'K1']
    })
df_r = pd.DataFrame({
        'B': ['B0', 'B1'],
        'key_r': ['K0', 'K1']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, left_on='key_l', right_on='key_r')
print(df_merged)
```

```python
# DataFrame merge on multiple keys (columns)
df_l = pd.DataFrame({
        'A': ['A0', 'A1'],
        'key1': ['K0', 'K1'],
        'key2': ['L0', 'L0']
    })
df_r = pd.DataFrame({
        'B': ['B1', 'B2'],
        'key1': ['K0', 'K0'],
        'key2': ['L0', 'L1']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, on=['key1', 'key2'], how='outer')
print(df_merged)
```

### Column Names

```python
# Same column name
df_l = pd.DataFrame({
        'A': ['L0', 'L1'],
        'key': ['K0', 'K1']
    })
df_r = pd.DataFrame({
        'A': ['R2', 'R3'],
        'key': ['K0', 'K1']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, on='key')
print(df_merged)
```

```python
# Add suffixes to the same column name
df_l = pd.DataFrame({
        'A': ['L0', 'L1'],
        'key': ['K0', 'K1']
    })
df_r = pd.DataFrame({
        'A': ['R2', 'R3'],
        'key': ['K0', 'K1']
    })
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, on='key', suffixes=('_l', '_r'))
print(df_merged)
```

### Merge on Index

```python
# Merge on index
df_l = pd.DataFrame({
        'A': ['A0', 'A1']
    }, index=['K0', 'K1'])
df_r = pd.DataFrame({
        'B': ['B0', 'B1']
    }, index=['K0', 'K1'])
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, left_index=True, right_index=True)
print(df_merged)
```

```python
# Merge on index and key
df_l = pd.DataFrame({
        'A': ['A0', 'A1'],
        'key': ['K0', 'K1']
    })
df_r = pd.DataFrame({
        'B': ['B0', 'B1']
    }, index=['K0', 'K1'])
print(df_l)
print(df_r)
df_merged = pd.merge(df_l, df_r, left_on='key', right_index=True)
print(df_merged)
```

## Concatenating DataFrames Along an Axis

### Index

```python
Image("images/concat-axis-0.png", width=272)
```

```python
# Keeping original indexes
df1 = pd.DataFrame({
        'A': ['A0', 'A1'],
        'B': ['B0', 'B1']
    }, index=['a', 'b'])
df2 = pd.DataFrame({
        'A': ['A2', 'A3'],
        'B': ['B2', 'B3']
    }, index=['c', 'd'])
print(df1)
print(df2)
df_concat = pd.concat([df1, df2])
print(df_concat)
```

```python
# Ignoring original indexes
df1 = pd.DataFrame({
        'A': ['A0', 'A1'],
        'B': ['B0', 'B1']
    }, index=['a', 'b'])
df2 = pd.DataFrame({
        'A': ['A2', 'A3'],
        'B': ['B2', 'B3']
    }, index=['c', 'd'])
print(df1)
print(df2)
df_concat = pd.concat([df1, df2], ignore_index=True)
print(df_concat)
```

```python
# Hierarchical index on the concatenation axis
df1 = pd.DataFrame({
        'A': ['A0', 'A1'],
        'B': ['B0', 'B1']
    }, index=['a', 'b'])
df2 = pd.DataFrame({
        'A': ['A2', 'A3'],
        'B': ['B2', 'B3']
    }, index=['c', 'd'])
print(df1)
print(df2)
df_concat = pd.concat([df1, df2], keys=['x', 'y'])
print(df_concat)
```

### Concatenating Axis

```python
Image("images/concat-axis-1.png", width=332)
```

```python
df1 = pd.DataFrame({
        'A': ['A0', 'A1'],
        'B': ['B0', 'B1']
    })
df2 = pd.DataFrame({
        'C': ['C0', 'C1'],
        'D': ['D0', 'D1']
    })
print(df1)
print(df2)
df_concat = pd.concat([df1, df2], axis=1)
print(df_concat)
```

### Logic on the Other Axis

```python
Image("images/concat-inner.png", width=242)
```

```python
# Inner join
df1 = pd.DataFrame({
        'A': ['A0', 'A1'],
        'B': ['B0', 'B1']
    }, index=[0, 1])
df2 = pd.DataFrame({
        'B': ['B2', 'B3'],
        'C': ['C2', 'C3']
    }, index=[2, 3])
print(df1)
print(df2)
df_concat = pd.concat([df1, df2], join='inner')
print(df_concat)
```

```python
Image("images/concat-outer.png", width=302)
```

```python
# Outer join
df1 = pd.DataFrame({
        'A': ['A0', 'A1'],
        'B': ['B0', 'B1']
    }, index=[0, 1])
df2 = pd.DataFrame({
        'B': ['B2', 'B3'],
        'C': ['C2', 'C3']
    }, index=[2, 3])
print(df1)
print(df2)
df_concat = pd.concat([df1, df2]) # default join='outer'
print(df_concat)
```

```python
Image("images/concat-join-axes.png", width=272)
```

```python
# Specifying indexes instead of inner/outer set logic
df1 = pd.DataFrame({
        'A': ['A0', 'A1'],
        'B': ['B0', 'B1']
    }, index=[0, 1])
df2 = pd.DataFrame({
        'B': ['B2', 'B3'],
        'C': ['C2', 'C3']
    }, index=[2, 3])
print(df1)
print(df2)
df_concat = pd.concat([df1, df2], join_axes=[df1.columns])
print(df_concat)
```
