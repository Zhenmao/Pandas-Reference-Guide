## DataFrame Index, Columns and Values

### Index (Row Index)

```python
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df.index)
print(df.index.values)
```

### Columns (Column Index)

```python
# DataFrame columns
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df.columns)
print(df.columns.values)
```

### Reindexing

```python
# Reindexing rows
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
df_reindexed = df.reindex(index=['row_2', 'row_1', 'row_0'])
print(df_reindexed)
```

```python
# Reindexing columns
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
df_reindexed = df.reindex(columns=['col_1', 'col_0'])
print(df_reindexed)
```

### Values

```python
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df.values)
```

### Transpose

```python
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df.T)
```

### With [ ]

```python
# df[col_label] selects columns
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df['col_0']) 
```

```python
# Convenience 1: [] with a slice to slice rows
# df[row_position_slice]
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df[1:2])
```

```python
# Convenience 2: [] with a boolean array to filter rows
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df['col_1']>10) # Returns a boolean array
print(df[df['col_0']>10])
```

```python
# Convenience 3: [] with a boolean DataFrame to filter element
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df>10)
print(df[df>10])
# This is used to set values elementwise based on some criterion
df[df>10] = 999
print(df)
```

## DataFrame Sorting and Ranking

### Sorting by Index

```python
# Sort rows by index
df = pd.DataFrame(np.arange(9).reshape((3,3)),
                  index=['row_2', 'row_1', 'row_0'],
                  columns=['col_2', 'col_1', 'col_0'])
print(df)
df_sorted = df.sort_index()
print(df_sorted)
```

```python
# Sort columns by index
df = pd.DataFrame(np.arange(9).reshape((3,3)),
                  index=['row_2', 'row_1', 'row_0'],
                  columns=['col_2', 'col_1', 'col_0'])
print(df)
df_sorted = df.sort_index(axis=1)
print(df_sorted)
```

### Sorting by Values

```python
# Sort rows by values
df = pd.DataFrame(np.arange(9)[::-1].reshape((3,3)),
                  index=['row_0', 'row_1', 'row_2'],
                  columns=['col_0', 'col_1', 'col_2'])
print(df)
df_sorted = df.sort_values(by=['col_0'])
print(df_sorted)
```

```python
# Sort columns by values
df = pd.DataFrame(np.arange(9)[::-1].reshape((3,3)),
                  index=['row_0', 'row_1', 'row_2'],
                  columns=['col_0', 'col_1', 'col_2'])
print(df)
df_sorted = df.sort_values(by=['row_0'], axis=1)
print(df_sorted)
```

### Ranking

```python
df = pd.DataFrame([
    [3, 4],
    [2, 5],
    [1, 4],
    [0, 5]
])
print(df)
print(df.rank(method='min'))
```

Tie-Breaking Methods with Rank

| Method  | Detail                                                    |
|:--------|:----------------------------------------------------------|
| Average | Average rank of group                                     |
| Min     | Lowest rank in group                                      |
| Max     | Highest rank in group                                     |
| First   | Ranks assigned in order they appear in the array          |
| Dense   | Like "min", but rank always increases by 1 between groups |

## DataFrame Computations

```python
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df.max())
```

```python
df = pd.DataFrame({
        'col_0': [0, 10, 20],
        'col_1': [1, 11, 21]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df.min(axis=1))
```

## DataFrame Missing Values

### Counting

```python
# Count the number of null values
df = pd.DataFrame({
        'col_0': [0, 10, np.nan],
        'col_1': [1, np.nan, np.nan]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df.count())
```

### Filtering

```python
# Whether the values are null
df = pd.DataFrame({
        'col_0': [0, 10, np.nan],
        'col_1': [1, np.nan, np.nan]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df.isnull())
```

```python
# Whether the values are not null
df = pd.DataFrame({
        'col_0': [0, 10, np.nan],
        'col_1': [1, np.nan, np.nan]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
print(df.notnull())
```

### Dropping

```python
# Drop rows with ANY null values
df = pd.DataFrame({
        'col_0': [0, 10, np.nan],
        'col_1': [1, np.nan, np.nan]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
df_dropped = df.dropna()
print(df_dropped)
```

```python
# Drop rows with ALL null values
df = pd.DataFrame({
        'col_0': [0, 10, np.nan],
        'col_1': [1, np.nan, np.nan]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
df_dropped = df.dropna(how='all')
print(df_dropped)
```

### Filling

```python
# Fill the null values
df = pd.DataFrame({
        'col_0': [0, 10, np.nan],
        'col_1': [1, np.nan, np.nan]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
df_filled = df.fillna(999)
print(df_filled)
```

```python
# Fill the null values, column specific
df = pd.DataFrame({
        'col_0': [0, 10, np.nan],
        'col_1': [1, np.nan, np.nan]
    }, index = ['row_0', 'row_1', 'row_2'])
print(df)
df_filled = df.fillna({'col_0': 777, 'col_1': 888})
print(df_filled)
```
